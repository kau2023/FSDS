Q1. Create an assert statement that throws an AssertionError if the variable spam is a negative
integer.
Sol:
Here's an example assert statement that throws an AssertionError if the variable spam is a negative integer:
assert spam >= 0, 'spam must be a non-negative integer'
This statement checks if the value of spam is greater than or equal to zero. If spam is a negative integer, the assertion will fail and raise an AssertionError with the message "spam must be a non-negative integer".

Q4. What are the two lines that must be present in your software in order to call logging.debug()?
Sol:
To use the logging.debug() method in your software, you need to import the logging module and configure the logging system. Here are the two lines that must be present:
import logging
logging.basicConfig(level=logging.DEBUG)

The first line imports the logging module, which provides the debug() method and other logging functionality. The second line sets up the logging system by configuring the root logger to output messages with a log level of DEBUG.

Once these two lines have been executed, you can call logging.debug() anywhere in your code to output debug messages to the console or other logging destinations. Additionally, you can customize the logging configuration by specifying a file to write the logs to, formatting options, and other settings.

Q6. What are the five levels of logging?
Sol:
The five levels of logging, in increasing order of severity, are:

DEBUG: Detailed information, typically of interest only when diagnosing problems.
INFO: Confirmation that things are working as expected.
WARNING: An indication that something unexpected or potentially problematic has happened.
ERROR: An indication that something has gone wrong and the software cannot perform a function.
CRITICAL: The highest level of severity, indicating a serious error that may cause the program to terminate.
Q7. What line of code would you add to your software to disable all logging messages?
Sol:
To disable all logging messages in your software, you can add the following line of code:
logging.disable(logging.CRITICAL)
This will disable all logging messages at or below the CRITICAL level.
Q8.Why is using logging messages better than using print() to display the same message?

Sol:
Using logging messages is often better than using print() to display the same message for several reasons:

Flexibility: With logging, you can control the level of logging messages that are displayed, which makes it easy to adjust the amount of detail in your program's output.

Configurability: Logging allows you to easily configure where the log messages are sent (e.g., console, file, syslog), the format of the messages, and other settings.

Granularity: By using different logging levels, you can log messages at different levels of granularity, such as DEBUG, INFO, WARNING, ERROR, and CRITICAL. This allows you to easily filter messages based on their level of importance.

Persistence: Logging messages can be stored to a file or a database, which allows you to review them at a later time, even after the program has completed execution.

Compatibility: Since logging is a standard library module, it is guaranteed to be available on any system running Python. This makes it more reliable than relying on a third-party library or tool for printing messages.
Q9.What are the differences between the Step Over, Step In, and Step Out buttons in the debugger?
Sol:
The Step Over, Step In, and Step Out buttons are all used to navigate through code when debugging, but they do so in slightly different ways:

Step Over: This button will execute the current line of code, and then move to the next line of code. If the current line of code contains a function call, the function will be executed, but the debugger will not enter the function's code. This is useful for quickly moving through code and only stopping at breakpoints or when an exception is raised.

Step In: This button will execute the current line of code, but if the current line contains a function call, the debugger will enter the called function and begin debugging that code. This is useful for stepping through code and seeing how functions are being executed.

Step Out: This button will execute the rest of the current function and return to the calling function. This is useful if you have stepped into a function and want to quickly return to the calling code without stepping through the rest of the function's code.
